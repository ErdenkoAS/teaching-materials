## Раздел 1. Командная строка

### Что такое CLI
**CLI (Command Line Interface)** это текстовый интерфейс взаимодействия с операционной системой. В отличие от GUI (графического интерфейса), вы вводите команды текстом и получаете текстовый ответ.

**Почему это важно для разработчика:**
- многие инструменты разработки (Git, npm, Docker, Python) изначально консольные;
- работа на серверах почти всегда происходит без GUI;
- консольные операции скриптуются и автоматизируются;
- быстрее, чем кликать мышью, если знать команды.

### Терминал, оболочка, командная строка

- **Терминал** программа-окно, в которой вы пишете команды (Terminal на macOS, Windows Terminal, GNOME Terminal).
- **Оболочка (shell)** программа, которая интерпретирует команды: `bash`, `zsh`, `PowerShell`, `cmd`.
- На Linux/macOS чаще всего `bash` или `zsh`. На Windows исторически `cmd`, сейчас рекомендуется `PowerShell` или `Git Bash`.

### Пути: абсолютные и относительные
- **Абсолютный путь** от корня файловой системы: `/home/user/project` или `C:\\Users\\user\\project`.
- **Относительный путь** от текущей директории: `./src/main.py`, `../docs`.
- `.` текущая директория, `..` родительская, `~` домашняя (Unix).

### Справочник команд Bash

Этот блок не нужно зубрить, держите его как шпаргалку. Главное освоить базовый набор команд (`pwd`, `ls`, `cd`, `mkdir`, `touch`, `cat`, `echo`, `cp`, `mv`, `rm`, `clear`, `man`). Удобный приём для запоминания - почти у каждой команды есть расшифровка.

#### Навигация по файловой системе

- `pwd` **P**rint **W**orking **D**irectory показать текущую директорию.
- `ls` **L**i**S**t список файлов. Полезные флаги: `ls -l` (подробно), `ls -a` (включая скрытые), `ls -la` (всё вместе).
- `cd` **C**hange **D**irectory перейти в директорию. `cd ..` подняться выше, `cd ~` в домашнюю, `cd /` в корень.

#### Работа с файлами и папками

- `mkdir` **M**a**K**e **DIR**ectory создать папку. Флаг `-p` создаёт промежуточные директории: `mkdir -p a/b/c`.
- `touch` создать пустой файл (или обновить дату модификации существующего).
- `cp` **C**o**P**y копировать. Для папок нужен `-r` (recursive).
- `mv` **M**o**V**e переместить или переименовать.
- `rm` **R**e**M**ove удалить. Для папок `-r`. **Осторожно: корзины нет!**

#### Просмотр содержимого

- `cat` con**CAT**enate вывести весь файл.
- `head -n 5 file.txt` первые 5 строк.
- `tail -n 5 file.txt` последние 5 строк.

#### Запись в файлы

- `echo "text"` вывести текст.
- `echo "text" > file.txt` записать в файл (перезапись).
- `echo "text" >> file.txt` дописать в конец.

#### Поиск и фильтрация

- `grep "pattern" file.txt` найти строки с pattern. Флаги: `-i` без учёта регистра, `-n` показать номера строк, `-v` инвертировать (строки без pattern).
- `find . -name "*.py"` найти все Python-файлы в текущей папке и подпапках.
- `wc -l file.txt` посчитать строки в файле.

#### Комбинирование команд

- `|` (pipe, конвейер) передать вывод одной команды на вход другой: `ls -la | grep ".py"`.
- `&&` выполнить следующую команду, **если предыдущая успешна**: `mkdir folder && cd folder`.
- `;` выполнить команды последовательно, независимо от результата.

#### Спецсимволы

- `*` любое количество любых символов (wildcard): `rm *.tmp` удалит все `.tmp`-файлы.
- `?` ровно один любой символ.
- `~` домашняя директория. `.` текущая. `..` родительская.

### Аналоги для Windows `cmd`

| Unix | Windows cmd |
|------|-------------|
| `ls` | `dir` |
| `cp` | `copy` |
| `mv` | `move` |
| `rm` | `del` |
| `cat` | `type` |
| `clear` | `cls` |

> Если установлен **Git Bash** (идёт вместе с Git for Windows), команды Unix работают и под Windows.


### Выполнение команд в Jupyter: `!` vs `%`

В Jupyter (и Google Colab) можно выполнять команды прямо из ячеек двумя способами, и **между ними есть важная разница**.

#### `!` запуск shell-команды

```
!pwd
!ls -la
!pip install pandas
```

Каждая команда с `!` выполняется в **новом подпроцессе**. Это значит, что **состояние не сохраняется** между ячейками: если вы сделаете `!cd folder`, в следующей ячейке вы снова окажетесь в исходной директории.

#### `%` magic-команды IPython

```
%cd folder      # перейти в папку (состояние сохраняется!)
%pwd            # показать текущую директорию
%ls             # список файлов
```

Magic-команды выполняются **в самом ядре Jupyter**, поэтому `%cd` действительно меняет рабочую директорию для всех последующих ячеек.

#### `%%writefile` записать содержимое ячейки в файл

```
%%writefile hello.py
print("Hello, World!")
```

Эта magic-команда удобнее, чем многократный `echo`, когда нужно записать многострочный файл (например, Python-скрипт).

#### Когда что использовать
- `%cd` для смены директории внутри ноутбука.
- `!` для всех остальных команд (создание файлов, git, pip и т.д.).
- `%%writefile` для записи многострочных файлов.

> Это особенность Jupyter. В реальном терминале никаких `!` и `%` нет, вы просто пишете команду.

## Раздел 3. Git

### Зачем нужна система контроля версий
Когда вы пишете курсовую или код, у вас часто появляются файлы вида:
```
diplom_final.docx
diplom_final_v2.docx
diplom_final_v2_итог_правки_научрука.docx
```

Это и есть «ручной контроль версий», и он плох:
- невозможно понять, что изменилось между версиями;
- если работает несколько человек, файлы конфликтуют;
- нельзя откатиться к произвольной точке.

**Git** распределённая система контроля версий, которая:
- хранит **всю историю изменений** проекта;
- позволяет **возвращаться к любой версии**;
- даёт работать **параллельно нескольким разработчикам**;
- хранит копию репозитория **локально** у каждого участника.

### Ключевые понятия

| Термин | Что это |
|--------|---------|
| **Репозиторий (repository)** | хранилище проекта вместе с историей |
| **Коммит (commit)** | зафиксированный «снимок» состояния проекта с описанием |
| **Рабочая директория** | файлы, которые вы реально видите и редактируете |
| **Индекс (staging area, index)** | «корзина», куда вы кладёте файлы перед коммитом |
| **HEAD** | указатель на текущий коммит |
| **Ветка (branch)** | независимая линия разработки |
| **Удалённый репозиторий (remote)** | копия проекта на сервере (например, GitHub) |

### Три состояния файла в Git

```
[Рабочая директория]  --git add-->  [Staging Area]  --git commit-->  [Репозиторий]
       (изменено)                       (индекс)                       (история)
```

### Минимальный набор команд

| Команда | Что делает |
|---------|-----------|
| `git init` | создать новый репозиторий в текущей папке |
| `git status` | посмотреть состояние файлов |
| `git add <file>` | добавить файл в индекс |
| `git add .` | добавить всё в индекс |
| `git commit -m "msg"` | зафиксировать изменения с сообщением |
| `git log` | посмотреть историю коммитов |
| `git diff` | посмотреть, что изменилось |

### Настройка Git (выполняется один раз)

```bash

```

Эти данные подписывают каждый ваш коммит.

## Раздел 5. Удалённый репозиторий, `.gitignore` и `README.md`

### Удалённые репозитории
**Remote-репозиторий** копия проекта на сервере. Самые популярные хостинги:
- **GitHub** (github.com) самый популярный, де-факто стандарт;
- **GitLab** часто используется в компаниях, есть self-hosted версия;
- **Bitbucket** интегрирован с экосистемой Atlassian.

### Связь локального и удалённого репозитория

```
[Локальный репозиторий] --git push--> [Удалённый (GitHub)]
[Локальный репозиторий] <--git pull-- [Удалённый (GitHub)]
```

### Основные команды

| Команда | Что делает |
|---------|-----------|
| `git clone <url>` | склонировать удалённый репозиторий к себе |
| `git remote add origin <url>` | привязать локальный репозиторий к удалённому |
| `git remote -v` | посмотреть список remote-ов |
| `git push -u origin main` | отправить ветку `main` на сервер (первый раз) |
| `git push` | отправить новые коммиты на сервер |
| `git pull` | забрать изменения с сервера |
| `git fetch` | скачать изменения без слияния |

### Типичный сценарий «опубликовать локальный проект на GitHub»

> Добавить блок с цитатой



1. Создать пустой репозиторий на GitHub (без README).
2. Скопировать его URL.
3. Локально выполнить:
   ```bash
   git remote add origin https://github.com/username/repo.git
   git branch -M main
   git push -u origin main
   ```

> На семинаре `push` мы делать **не будем** (требуется аккаунт и аутентификация), но команды нужно знать.

### Файл `.gitignore`

Не всё нужно хранить в репозитории. **Игнорируем:**
- скомпилированные файлы (`*.pyc`, `__pycache__/`);
- виртуальные окружения (`venv/`, `.venv/`);
- IDE-настройки (`.idea/`, `.vscode/`);
- секреты (`.env`, ключи API);
- логи и временные файлы (`*.log`, `*.tmp`);
- большие данные и обученные модели (`*.csv` больших размеров, `*.pkl`).

Пример `.gitignore` для Python-проекта:
```
# Python
__pycache__/
*.pyc
*.pyo
venv/
.venv/

# IDE
.idea/
.vscode/

# OS
.DS_Store
Thumbs.db

# Secrets
.env
*.key
```

### Файл `README.md`

**README** «лицо» репозитория. Это первое, что видит человек, открывая проект на GitHub. Минимальное содержание:

```markdown
# Название проекта

Краткое описание в одно-два предложения.

## Установка
Как развернуть проект локально.

## Использование
Примеры запуска.

## Автор
Имя, контакт.
```

Формат **Markdown** простая текстовая разметка: `#` заголовок, `**жирный**`, `*курсив*`, `` `код` ``, `- ` пункт списка.
